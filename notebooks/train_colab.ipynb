```python
# Update data_root to point at local copies
import yaml

DATASET = 'ravdess'  # change to 'esd' for ESD-English
CONFIG_PATH = f'configs/crnn_{DATASET}.yaml'

with open(CONFIG_PATH) as f:
    cfg = yaml.safe_load(f)

cfg['data_root'] = f'{LOCAL_DATA}/{DATASET}'

runtime_config = f'/content/crnn_{DATASET}_runtime.yaml'
with open(runtime_config, 'w') as f:
    yaml.dump(cfg, f)

print(f'Using data_root: {cfg["data_root"]}')
```

```python
!python train.py --config {runtime_config}
```
